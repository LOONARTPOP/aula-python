```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados_df = pd.read_csv("dados_climaticos.csv", sep=";", encoding='utf-8', low_memory=False)
```

```python
#Tratamento de dados -> necessario para que o python entenda o que esta os tipos de dados que iremos utilizar
dados_df["Data"] = pd.to_datetime(dados_df["Data"], dayfirst=True, errors='coerce')
display(dados_df)
```

```python
# Lista de colunas numÃ©ricas
colunas_numericas = [
"Temp. Ins. (C)", "Temp. Max. (C)", "Temp. Min. (C)", "Umi. Ins. (%)",
"Umi. Max. (%)", "Umi. Min. (%)", "Pto Orvalho Ins. (C)", "Pto Orvalho Max. (C)",
"Pto Orvalho Min. (C)", "Pressao Ins. (hPa)", "Pressao Max. (hPa)",
"Pressao Min. (hPa)", "Vel. Vento (m/s)", "Dir. Vento (m/s)", "Raj. Vento (m/s)",
"Radiacao (KJ/m²)", "Chuva (mm)"]
```

```python
for col in colunas_numericas:
    #Troca virgula pelo ponto
    dados_df[col] = dados_df[col].astype(str).str.replace(',', '.', regex=False)

    #Converte a string que representa valores reais em numéricos (float)
    dados_df[col] = pd.to_numeric(dados_df[col], errors='coerce')
```

```python
#configura estilo do grafico
sns.set(style="whitegrid")
#cria uma paleta de cores com o estilo "husl", com uma cor diferente para cada coluna numerica.
colors = sns.color_palette("husl", len(colunas_numericas))

#cria uma figura com varios subgraficos(subplots) dispostos em uma coluna (1 por linha)
#o figsize ajusta o tamanho da imagem final proporcionalmente ao numero de graficos
fig, axes = plt.subplots(len(colunas_numericas), 1, figsize=(12,5*len(colunas_numericas)))

for i, col in enumerate(colunas_numericas):
    #Gera grafico de linha (lineplot) para a variavel col ao longo da coluna de datas(dadosdfdatas)
    #usa um dos eixos (axes[i]) e aplica uma cor distinta
    #adiciona marcado em cada ponto da linha (marker=o)
    sns.lineplot(x=dados_df["Data"], y=dados_df[col], ax=axes[i], color=colors[i], marker="o")
    #define o titulo e rotulos dos eixos
    axes[i].set_title(f"{col} ao longo do Tempo")
    axes[i].set_xlabel("Data")
    axes[i].set_ylabel(col)
    #Gira os rotulos do eixo x para 45º para melhorar a leitura
    axes[i].tick_params(axis='x', rotation=45)

    #ajusta o layout para evitar sobreposicao
    plt.tight_layout()
    #mostra grafico
    plt.show()
    
display(dados_df)
```
